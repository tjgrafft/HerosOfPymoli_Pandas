### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
purchase_data["SN"].value_counts()
player_count = len(purchase_data["SN"].value_counts())
player_breakdown = pd.DataFrame({"Total Players": [player_count]})
player_breakdown
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Unique items variable
unique_items = len(purchase_data["Item ID"].unique())

#Avg price variable
avg_price = purchase_data["Price"].mean()

#Total Purchase variable
total_purchase = len(purchase_data["Purchase ID"])

#Total Revenue variable
total_rev = purchase_data["Price"].sum()

#Creating new dataframe
purchase_breakdown = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [avg_price],
                                  "Total Purchases": [total_purchase], "Total Revenue": [total_rev]})
#Formatting new table
purchase_breakdown = purchase_breakdown.round(2)
purchase_breakdown["Average Price"] = purchase_breakdown["Average Price"].map("${:.2f}".format)
purchase_breakdown["Total Revenue"] = purchase_breakdown["Total Revenue"].map("${0:,.2f}".format)
purchase_breakdown
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Create new table for gender
gender_breakdown = purchase_data[["SN", "Gender"]]
gender_breakdown = gender_breakdown.drop_duplicates(subset='SN', keep='first')

#Total Gender
total_gender = gender_breakdown["Gender"].count()

#Create variable for only male gender
male = gender_breakdown["Gender"].value_counts()['Male']
male_percent = (male/total_gender) * 100
#Create variable for only female gender
female = gender_breakdown["Gender"].value_counts()['Female']
female_percent = (female/total_gender) * 100

#Create variable for Other/Non-disclosed
other = total_gender - male - female
other_percent = (other/total_gender) * 100

#Create new dataframe for gender
gender_table = pd.DataFrame(gender_breakdown["Gender"].value_counts())
percent_table = pd.DataFrame({"Percentage of Players": [male_percent, female_percent, other_percent]})
percent_table = percent_table.round(2)
percent = [male_percent, female_percent, other_percent]
gender_table["Percentage of Players"] = percent
gender_table.round(2)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group by Genders
purchase_group = purchase_data.groupby(["Gender"])

#Average Price Per Gender
gender_price = purchase_group[["Price"]].mean()
gender_price = gender_price.round(2)
gender_price

#Purchase Count per Gender
gender_count = purchase_group[["Purchase ID"]].count()
gender_count

#Total Purchase Price per Gender
gender_total_value = purchase_group[["Price"]].sum()
gender_total_value


#Merge Table
merge_table = pd.merge(gender_count, gender_price, on="Gender")
merge_table["Total Purchase Value"] = gender_total_value

# Avg Total Purchase Per Person
gender_average_total = (merge_table["Total Purchase Value"] / gender_table["Gender"])
gender_average_total = gender_average_total.round(2)

merge_table = merge_table.rename(columns={"Purchase ID":"Purchase Count", "Price":"Average Purchase Price"})

merge_table["Avg Total Value Per Person"] = gender_average_total
merge_table

merge_table["Average Purchase Price"] = merge_table["Average Purchase Price"].map("${:.2f}".format)
merge_table["Total Purchase Value"] = merge_table["Total Purchase Value"].map("${0:,.2f}".format)
merge_table["Avg Total Value Per Person"] = merge_table["Avg Total Value Per Person"].map("${:.2f}".format)
merge_table
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held  
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]

# Create the names for the four bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


#Create new table for age
age_breakdown = purchase_data[["SN", "Age"]]
age_breakdown = age_breakdown.drop_duplicates(subset='SN', keep='first')
pd.cut(age_breakdown["Age"], bins, labels=group_names).head()
age_breakdown["Age_test"] = pd.cut(age_breakdown["Age"], bins, labels=group_names)
age_group = age_breakdown.groupby("Age_test").count()
age_group = age_group.rename(columns={"SN":"Total Count", "Age":"Percentage of Players"})


#Age tables

total_age = age_breakdown["Age"].count()

percent_df = (age_group["Total Count"] / total_age) * 100

age_group["Percentage of Players"] = percent_df
age_group = age_group.round(2)
age_group
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group by Age

age_purchase_group = purchase_data[["Purchase ID", "Age", "Price"]]
age_purchase_group["Age_test"] = pd.cut(age_purchase_group["Age"], bins, labels=group_names)

age_purchase_df = age_purchase_group.groupby("Age_test")
age_purchase_df

#Average Price Per Gender
age_price = age_purchase_df[["Price"]].mean()
age_price = age_price.round(2)
age_price

#Purchase Count per Gender
age_count = age_purchase_df[["Purchase ID"]].count()
age_count

#Total Purchase Price per Gender
age_total_value = age_purchase_df[["Price"]].sum()
age_total_value

#merge age
merge_age = pd.merge(age_count, age_price, on="Age_test")
merge_age["Total Purchase Value"] = age_total_value

#Average Total Purchase Per Person
average_totalper = (merge_age["Total Purchase Value"] / age_group["Total Count"])
average_totalper = average_totalper.round(2)

merge_age = merge_age.rename(columns={"Purchase ID":"Purchase Count", "Price":"Average Purchase Price"})

merge_age["Avg Total Value Per Person"] = average_totalper
merge_age

merge_age["Average Purchase Price"] = merge_age["Average Purchase Price"].map("${:.2f}".format)
merge_age["Total Purchase Value"] = merge_age["Total Purchase Value"].map("${0:,.2f}".format)
merge_age["Avg Total Value Per Person"] = merge_age["Avg Total Value Per Person"].map("${:.2f}".format)
merge_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_df = purchase_data.groupby('SN')['Price'].agg(['count', 'mean', 'sum'])
largest_spender = spenders_df.nlargest(5, 'sum')
new_round = largest_spender['mean'].round(2)

largest_spender = largest_spender.rename(columns={"count":"Purchase Count", "mean":"Average Purchase Price", "sum":"Total Purchase Value"})
largest_spender["Average Purchase Price"] = new_round
largest_spender
#group by SN then sum total and then get largest

largest_spender["Average Purchase Price"] = largest_spender["Average Purchase Price"].map("${:.2f}".format)
largest_spender["Total Purchase Value"] = largest_spender["Total Purchase Value"].map("${0:,.2f}".format)

largest_spender
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_item_df = purchase_data[["Item ID", "Item Name", "Price"]]

pop_df = pop_item_df.groupby(["Item ID", "Item Name"]).count()
pop_df = pop_df.nlargest(5, 'Price')


largest_pop = pop_df.rename(columns={"Price":"Purchase Count"})

largest_pop
```


## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
prof_item_df = purchase_data[["Item ID", "Item Name","Price"]]
prof_df = prof_item_df.groupby('Item ID')['Price'].agg(['count', 'mean', 'sum'])
largest_prof = prof_df.nlargest(5, 'sum')

largest_prof = largest_prof.rename(columns={"count":"Purchase Count", "mean":"Total Price", "sum":"Total Purchase Value"})
largest_prof

largest_prof["Total Price"] = largest_prof["Total Price"].map("${:.2f}".format)
largest_prof["Total Purchase Value"] = largest_prof["Total Purchase Value"].map("${0:,.2f}".format)
largest_prof
```
